```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Load the datasets
regression_1_path = r'C:\Users\aditya\Desktop\Homework 7\regression_1.csv'
regression_2_path = r'C:\Users\aditya\Desktop\Homework 7\regression_2.csv'
classification_path = r'C:\Users\aditya\Desktop\Homework 7\classification.csv'

# Read the data
regression_1 = pd.read_csv(regression_1_path)
regression_2 = pd.read_csv(regression_2_path)
classification = pd.read_csv(classification_path)

# Regression 1: Linear Regression
X1 = regression_1[['x1']]
y1 = regression_1['x2']
linear_model = LinearRegression()
linear_model.fit(X1, y1)

# Regression 2: Polynomial Regression
X2 = regression_2[['x1']]
y2 = regression_2['x2']
polynomial_model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
polynomial_model.fit(X2, y2)

# Classification: Logistic Regression
X3 = classification[['x1', 'x2']]
y3 = classification['label']
logistic_model = LogisticRegression()
logistic_model.fit(X3, y3)

# Visualization
plt.figure(figsize=(15, 5))

# Plot Regression 1
plt.subplot(1, 3, 1)
sns.scatterplot(data=regression_1, x="x1", y="x2", color="blue", alpha=0.7)
plt.plot(X1, linear_model.predict(X1), color="red", label="Fitted Line")
plt.title("Regression 1: Linear Regression")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plot Regression 2
plt.subplot(1, 3, 2)
sns.scatterplot(data=regression_2, x="x1", y="x2", color="green", alpha=0.7)
x_range = np.linspace(X2.min(), X2.max(), 100).reshape(-1, 1)
plt.plot(x_range, polynomial_model.predict(x_range), color="red", label="Fitted Curve")
plt.title("Regression 2: Polynomial Regression")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plot Classification
plt.subplot(1, 3, 3)
sns.scatterplot(data=classification, x="x1", y="x2")
xx, yy = np.meshgrid(
    np.linspace(X3['x1'].min(), X3['x1'].max(), 100),
    np.linspace(X3['x2'].min(), X3['x2'].max(), 100)
)
Z = logistic_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.2, levels=np.arange(-0.5, 2, 1), cmap="Set1")
plt.title("Classification: Logistic Regression")
plt.xlabel("x1")
plt.ylabel("x2")

plt.tight_layout()
plt.show()
```
