```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\MS\Germany\List of universities\FAU Erlangen\Semester 2\Data Science Survival Skills\Homework submission\Homework 4\census_income_dataset.csv'
data = pd.read_csv(file_path)

plt.rcParams.update({'font.size': 14})

plt.figure(figsize=(10, 6))
age_counts = data['AGE'].value_counts().sort_index()  # Count occurrences of each age
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-', color='purple')
plt.title('Age Distribution of Respondents')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.savefig(r'C:\MS\Germany\List of universities\FAU Erlangen\Semester 2\Data Science Survival Skills\Homework submission\Homework 4\Aditya\age.svg', format='svg')
plt.show()

relationship_counts = data['RELATIONSHIP'].value_counts()

plt.figure(figsize=(8, 8))
relationship_counts.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette('bright')  # Vibrant color palette
)
plt.title('Relationship Status Distribution')
plt.ylabel('')  # Hide y-axis label for cleaner look
plt.tight_layout()
plt.savefig(r'C:\MS\Germany\List of universities\FAU Erlangen\Semester 2\Data Science Survival Skills\Homework submission\Homework 4\Aditya\relationship.svg', format='svg')
plt.show()

salary_by_education = data.groupby(['EDUCATION', 'SALARY']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
salary_by_education.plot(
    kind='bar', 
    color=['skyblue', 'salmon'],  # Different colors for each category
    width=0.7, 
    figsize=(14, 8)
)
plt.title('Salary Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Salary', labels=['<=50K', '>50K'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(r'C:\MS\Germany\List of universities\FAU Erlangen\Semester 2\Data Science Survival Skills\Homework submission\Homework 4\Aditya\salary.svg', format='svg')
plt.show()
```
