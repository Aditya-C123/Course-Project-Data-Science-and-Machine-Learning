```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\vindu\Desktop\winequality-red.csv'  # Replace with your file path
wine_data = pd.read_csv(file_path)

# Prepare the data
features = wine_data.drop(columns=['quality'])
labels = wine_data['quality']

# Function to create scatter plots
def scatter_plot(data, title, labels, ax, cmap='viridis'):
    scatter = ax.scatter(data[:, 0], data[:, 1], c=labels, cmap=cmap, s=10, alpha=0.7)
    ax.set_title(title)
    ax.axis('off')
    return scatter

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(features)

# Apply UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
umap_result = umap_reducer.fit_transform(features)

# Create plots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# PCA Plot
scatter_plot(pca_result, "PCA Visualization", labels, axes[0])

# t-SNE Plot
scatter_plot(tsne_result, "t-SNE Visualization", labels, axes[1])

# UMAP Plot
scatter_plot(umap_result, "UMAP Visualization", labels, axes[2])

# Add a color bar for quality
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap='viridis'), ax=axes, orientation='horizontal', fraction=0.02, pad=0.1)
cbar.set_label('Quality')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway
import pandas as pd
file_path = r'C:\Users\vindu\Desktop\winequality-red.csv'
wine_data = pd.read_csv(file_path)
grouped_alcohol = [wine_data[wine_data['quality'] == q]['alcohol'] for q in wine_data['quality'].unique()]
f_stat, p_value = f_oneway(*grouped_alcohol)
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
alpha = 0.05 
if p_value < alpha:
    print("Reject the null hypothesis: Alcohol content significantly contributes to wine quality.")
else:
    print("Fail to reject the null hypothesis: Alcohol content does not significantly contribute to wine quality.")
```

```python
from scipy.stats import kruskal
import pandas as pd
file_path = r'C:\MS\Germany\List of universities\FAU Erlangen\Semester 2\Data Science Survival Skills\Homework submission\Homework 5\winequality-red.csv'
wine_data = pd.read_csv(file_path)
grouped_sugar = [wine_data[wine_data['quality'] == q]['residual sugar'] for q in wine_data['quality'].unique()]
h_stat, p_value = kruskal(*grouped_sugar)
print(f"H-statistic: {h_stat}")
print(f"P-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Residual sugar significantly contributes to wine quality.")
else:
    print("Accept the null hypothesis: Residual sugar does not significantly contribute to wine quality.")
```
